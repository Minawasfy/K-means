```python
df_setosa= df[df['Species'] =='Iris-setosa']
df_versicolor= df[df['Species'] =='Iris-versicolor']
df_virginica= df[df['Species'] =='Iris-virginica']

df_setosa
```

```python
ax = df_setosa.plot.scatter(x='SepalLengthCm', y='SepalWidthCm',  color='red', label='Iris - Setosa')
df_versicolor.plot.scatter(x='SepalLengthCm', y='SepalWidthCm',  color='blue', label='Iris-versicolor', ax = ax)
df_virginica.plot.scatter(x='SepalLengthCm', y='SepalWidthCm',  color='green', label='Iris-virginica',ax = ax)
```

```python
# Running KMeans for a range of clusters to apply the elbow method
wcss = []  # within-cluster sum of squares

for i in range(1, 11):  # trying clusters from 1 to 10
    # n_init = Number of times the k-means algorithm is run with different centroid seeds
    kmeans = KMeans(n_clusters=i, n_init=10, random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)  # inertia_ contains WCSS for a particular k

# Plotting the elbow method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1, 11, 1))
plt.grid(True)
plt.show()
```

```python
# Initializing KMeans with the number of clusters (in this case, 3)
kmeans = KMeans(n_clusters=3, n_init=10)

# Fitting the KMeans model to the standardized data
kmeans.fit(df)
```

```python
# Getting the cluster labels and centroids
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
cluster_labels
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1], c= cluster_labels)
```

```python
# Visualizing the clusters and centroids
plt.figure(figsize=(8, 6))

# Plotting data points with their assigned clusters
for i in range(len(df)):
    if cluster_labels[i] == 0:
        plt.scatter(df.iloc[i, 0], df.iloc[i, 1], color='red', s=50, label='Cluster 1' if i == 0 else "")
    elif cluster_labels[i] == 1:
        plt.scatter(df.iloc[i, 0], df.iloc[i, 1], color='blue', s=50, label='Cluster 2' if i == 0 else "")
    else:
        plt.scatter(df.iloc[i, 0], df.iloc[i, 1], color='green', s=50, label='Cluster 3' if i == 0 else "")

# Plotting centroids
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', s=100, label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2 ')
plt.legend()
plt.show()
```
